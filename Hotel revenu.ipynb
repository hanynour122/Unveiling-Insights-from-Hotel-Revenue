```python
df['month_number'] = pd.to_datetime(df['arrival_date_month'], format='%B').dt.month
df['month_number']=df['month_number'].astype(str)
```

```python
df["arrival_date_year"]=df["arrival_date_year"].astype(str)
df["arrival_date_day_of_month"]=df["arrival_date_day_of_month"].astype(str)
```

```python
df["date"] = df['month_number'] + '/' + df["arrival_date_day_of_month"]+ '/'+df["arrival_date_year"]
```

```python
df["date"]=pd.to_datetime(df["date"])
df["reservation_status_date"]=pd.to_datetime(df["reservation_status_date"])
```

```python
df["lead_time"]= df["date"] - df["reservation_status_date"]
```

```python
df[df["lead_time"].dt.days<0].count()
```

```python
def replace_year(date, old_year, new_year):
    if date.year == old_year:
        return date.replace(year=new_year)
    return date

old_year = 2020
new_year = 2019
df['reservation_status_date'] = df['reservation_status_date'].apply(lambda x: replace_year(x, old_year, new_year))
```

```python
df["lead_time"]= df["date"] - df["reservation_status_date"]
df[df["lead_time"].dt.days<0].count()
```

```python
plt.boxplot(df["adr"], vert=True)
          
# Add title and labels
plt.title('Box Plot of Lead Time')
plt.xlabel('Lead Time')

# Show the plot
plt.show()
```

```python
Q1 = df[df["adr"] > 0]["adr"].quantile(0.25)
Q3 = df[df["adr"] > 0]["adr"].quantile(0.75)

IQR = Q3 - Q1

lower = max(Q1 - IQR * 1.5, 0)
upper = Q3 + IQR * 1.5

outlier = df[(df["adr"] > upper) | (df["adr"] < lower)]
non_outlier = df[(df["adr"] < upper) | (df["adr"] > lower)]
```

```python
outlier["adr"].count()
```

```python
df.loc[df["adr"]>upper , "adr"]= upper
df.loc[df['adr'] < lower, 'adr'] = lower
```

```python
plt.boxplot(df["adr"], vert=True)
plt.title('Box Plot of Lead Time')
plt.xlabel('Lead Time')
plt.show()
```

```python
grouped_hotel = df['hotel'].groupby(df['hotel']).count().reset_index(name='counts')
grouped_hotel
```
